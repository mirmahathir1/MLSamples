```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from PIL import Image
import ast
```

```python
train_df = pd.read_csv("../input/train.csv")
```

```python
train_df.head(5)
```

```python
train_df.shape
```

```python
sample_index=1
im = np.array(Image.open('../input/train/'+train_df.loc[sample_index].image_id+'.jpg'), dtype=np.uint8)

# Create figure and axes
fig,ax = plt.subplots(1)

# Display the image
ax.imshow(im)

data = ast.literal_eval(train_df.loc[sample_index].bbox)

x=data[0]
y=data[1]
w=data[2]
h=data[3]


print(type(train_df.loc[sample_index].bbox))
# Create a Rectangle patch
rect = patches.Rectangle((x,y),w,h,linewidth=1,edgecolor='r',facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
```
